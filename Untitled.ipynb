10. For this question, you're going to create some visualizations to compare GDP per capita values for the years 1990, 2000, 2010, and 2020. Start by subsetting your data to just these 4 years into a new DataFrame named gdp_decades. Using this, create the following 4 plots:

A boxplot
A barplot (check out the Barplot with Seaborn section: https://www.python-graph-gallery.com/barplot/#Seaborn)
A scatterplot
A scatterplot with a trend line overlaid (see this regplot example: https://www.python-graph-gallery.com/42-custom-linear-regression-fit-seaborn)
Comment on what you observe has happened to GDP values over time and the relative strengths and weaknesses of each type of plot.

```python
gdp_decades = gdp[gdp['Year'].isin([1990, 2000, 2010, 2020])]
```

```python
gdp_decades
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Year', y='GDP_Per_Capita', data=gdp_decades)
plt.title('GDP per Capita: Boxplot for Selected Years')
plt.ylabel('GDP Per Capita')
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Year', y='GDP_Per_Capita', data=gdp_decades, estimator=np.mean)
plt.title('Average GDP per Capita: Selected Years')
plt.ylabel('Average GDP Per Capita')
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Year', y='GDP_Per_Capita', data=gdp_decades, jitter=True, alpha=0.6)
plt.title('Scatterplot of GDP per Capita for Selected Years')
plt.ylabel('GDP Per Capita')
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Year', y='GDP_Per_Capita', data=gdp_decades, scatter_kws={'alpha':0.6})
plt.title('Trend Line of GDP per Capita (1990, 2000, 2010, 2020)')
plt.ylabel('GDP Per Capita')
plt.xlabel('Year')
plt.show()
```

Which country had the highest GDP per capita in 2020? Create a plot showing how this country's GDP per capita has changed over the timespan of the dataset.

```python
# Subset for the year 2020
gdp_2020 = gdp[gdp['Year'] == 2020]

# Find the country with highest GDP per capita
highest_gdp_country_2020 = gdp_2020.loc[gdp_2020['GDP_Per_Capita'].idxmax()]['Country']
highest_gdp_value = gdp_2020['GDP_Per_Capita'].max()

print(f"Highest GDP per capita in 2020: {highest_gdp_country_2020} (${highest_gdp_value:.2f})")
```

```python
# Subset data for the country with highest GDP per capita
country_gdp_history = gdp[gdp['Country'] == highest_gdp_country_2020]

# Plot GDP per capita over time
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='GDP_Per_Capita', data=country_gdp_history, marker='o')
plt.title(f"GDP Per Capita Over Time: {highest_gdp_country_2020}")
plt.ylabel('GDP Per Capita')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

13. Which country had the lowest GDP per capita in 2020? Create a plot showing how this country's GDP per capita has changed over the timespan of the dataset. Bonus question: Is it true in general that coutries had a higher GDP per capita in 2020 than in 1990? Which countries had lower GDP per capita in 2020 than in 1990?

```python
# Filter the dataset for 2020
gdp_2020 = gdp[gdp['Year'] == 2020]

# Find the country with the lowest GDP per capita
lowest_gdp_country_2020 = gdp_2020.loc[gdp_2020['GDP_Per_Capita'].idxmin()]['Country']
lowest_gdp_value = gdp_2020['GDP_Per_Capita'].min()

print(f"Lowest GDP per capita in 2020: {lowest_gdp_country_2020} (${lowest_gdp_value:.2f})")
```

```python
# Subset data for the lowest GDP country
lowest_country_gdp_history = gdp[gdp['Country'] == lowest_gdp_country_2020]

# Plot the GDP per capita over time
plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='GDP_Per_Capita', data=lowest_country_gdp_history, marker='o')
plt.title(f"GDP Per Capita Over Time: {lowest_gdp_country_2020}")
plt.ylabel('GDP Per Capita')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```
